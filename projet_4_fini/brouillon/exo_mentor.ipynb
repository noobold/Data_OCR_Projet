```python
"""exo mentor
avec aide_alimentaire.csv convertir toutes les données en kilo au lieu de millier de tonne
en seulement 4 ligne fait ajout dico pour créer des nouvelle colonneset drop les anciennes"""

import pandas as pd
dispo=pd.read_csv('donnee/dispo_alimentaire.csv')
dispo.head()
```

### exo en 4 lignes

```python
"""debut exo fonction 2 ligne reste 2"""
def conv(valeur):
    return valeur *pow(10,6)

#test conv
#conv(123456)
#conv(123456,6)
```

```python
"""suite exo 2 ligne max"""

for a in range (9,18):
    dispo.iloc[:,a] = dispo.iloc[:,a].map(conv)
```

### avec dictionnaire

```python
#voir pour faire une boucle avec df.columns.tolist()

DicoColonne = {
    "Disponibilité intérieure" : "Disponibilité intérieure en kilo",
    "Exportations - Quantité" : "Exportations - Quantité en kilo",
    "Importations - Quantité" : "Importations - Quantité en kilo",
    "Nourriture" : "Nourriture en kilo",
    "Pertes" : "Pertes en kilo",
    "Production" : "Production en kilo",
    "Semences" : "Semences en kilo",
    "Traitement" : "Traitement en kilo",
    "Variation de stock" : "Variation de stock en kilo"
    }

for clé,valeur in DicoColonne.items():
    dispo[valeur] = dispo[clé]*10**6
    dispo.pop(clé)
```
